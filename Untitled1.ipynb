```python
brand_performnace = df.groupby('Description').agg({
    'TotalSalesDollars': 'sum',
    'ProfitMargin':'mean'    
}).reset_index()
```

```python
brand_performnace.head()
```

```python
low_sales_threshold =brand_performnace['TotalSalesDollars'].quantile(0.15)
high_margin_threshold = brand_performnace['ProfitMargin'].quantile(0.85)
```

```python
low_sales_threshold
```

```python
high_margin_threshold
```

```python
target_brands = brand_performnace[
    (brand_performnace['TotalSalesDollars']<low_sales_threshold) &
    (brand_performnace['ProfitMargin']>high_margin_threshold)
    
    ]
print("Brand with low sales but profit margin")
display(target_brands.sort_values('TotalSalesDollars'))
```

```python
plt.figure(figsize=(15,10))
sns.scatterplot(
    data=brand_performnace,
    x = 'TotalSalesDollars',
    y = 'ProfitMargin',
    color = "blue",
    label = "All Brands",
    alpha = 0.2

    
)

sns.scatterplot(

    data = target_brands,
    x = 'TotalSalesDollars',
    y = 'ProfitMargin',
    color = "Red",
    label = "Target Brand",
    alpha = 0.2   
)

plt.axhline(
    y = high_margin_threshold,
    linestyle = '--',
    color = "black",
    label = "high_margin_threshold"
    
)

plt.axvline(
    x = low_sales_threshold,
    linestyle = '--',
    color = "Black",
    label = "Low Sales threshold"
)

plt.xlabel("Total Sales ($)")
plt.ylabel("Profit Margin (%)")
plt.title("Brands for Promotional Pricing Adjustments")

plt.legend()
plt.grid(True)
plt.show()
```

```python
def format_dollars(value):
    if value >= 1_000_000:
        return f"{value/1_000_000:.2f}M"

    elif value >= 1_000:
        return f"{value/1_000:.2f}k"

    else:
        return str(value)
```

```python
vendor_perfomace = df.groupby('VendorName').agg({
    'TotalPurchaseDollars':'sum',
    'GrossProfit':'sum',
    'TotalSalesDollars':'sum'        
})
```

```python
vendor_perfomace
```

```python
vendor_perfomace['Purchase Contribution%']=vendor_perfomace['TotalPurchaseDollars']/vendor_perfomace['TotalPurchaseDollars'].sum()*100 
```

```python
vendor_perfomace['Purchase Contribution%']
```

```python
vendor_perfomace = round(vendor_perfomace.sort_values('Purchase Contribution%', ascending = False),2)
```

```python
vendor_perfomace 
```

```python
top_vendors = vendor_perfomace.head(10) 
top_vendors['TotalPurchaseDollars'] = vendor_perfomace['TotalPurchaseDollars'].apply(format_dollars)
top_vendors['GrossProfit'] = vendor_perfomace['GrossProfit'].apply(format_dollars)
top_vendors['TotalSalesDollars'] = vendor_perfomace['TotalSalesDollars'].apply(format_dollars)
top_vendors
```

```python
top_vendors['cumulative_contribution'] =  top_vendors['Purchase Contribution%'].cumsum()
```
